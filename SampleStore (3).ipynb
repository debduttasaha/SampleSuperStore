```python
c = df.corr()
sns.heatmap(c, annot=True, cmap='coolwarm')
plt.show()
```

From the correlation matrix we can the that there is a high correlation between sales and profit. Also we can a good correlation between quantity and sales

```python
plot = ["Ship Mode", "Segment", "Region", "Category", "Sub-Category"]
```

```python
fig = plt.figure(figsize=(12,10))
for i in plot:
    sns.countplot(x=i, data=df)
    plt.show()
```

# Analyzing the Sales of the Store with respect to categorical features

```python
sales_data = df.groupby('Category')[['Sales']].sum()
print(sales_data)

plt.pie(sales_data['Sales'], labels=sales_data.index)

plt.show()
```

```python
sales_data = df.groupby('Sub-Category')[['Sales']].sum()
print(sales_data)

plt.pie(sales_data['Sales'], labels=sales_data.index)

plt.show()
```

```python
sales_data = df.groupby('Region')[['Sales']].sum()
print(sales_data)

plt.pie(sales_data['Sales'], labels=sales_data.index)

plt.show()
```

```python
sales_data = df.groupby('Segment')[['Sales']].sum()
print(sales_data)

plt.pie(sales_data['Sales'], labels=sales_data.index)

plt.show()
```
